```python
import matplotlib.pyplot as plt
```

```python
# Vẽ biểu đồ: Line Chart
x = [1, 2, 3, 4, 5]
y = [10, 7, 15, 30, 6]

x1 = [1, 2, 3, 4, 5]
y1 = [25, 15, 20, 9, 25]

plt.plot(x, y, 'bo--', label = 'Dữ liệu')# Vẽ đường 1
plt.plot(x1, y1, marker = 'o', color = 'r', linestyle = '-', label = 'Dữ liệu-1')# Vẽ đường 2
plt.legend()#hiển thị chú giải
plt.xlabel('Trục x')#đặt nhãn cho trục x
plt.xlabel('Trục y')#đặt nhãn cho trục y
plt.show()#show biểu đồ
```

```python
# Vẽ biểu đồ cột Bar Chart
catergories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 20, 16, 25]

plt.bar(catergories, values, color = 'b', alpha = 0.5)#vẽ biểu đồ cột
plt.xlabel('Danh mục')#đặt nhãn cho trục x
plt.ylabel('Giá trị')#đặt nhãn cho trục y
plt.show()
```

```python
# Vẽ biểu đồ hình tròn Pie Chart
my_labels = ['A', 'B', 'C', 'D']
sizes = [10, 15, 20, 45]
my_colors = ['blue', 'red', 'yellow', 'green']

plt.pie(sizes, labels = my_labels, colors = my_colors, autopct='%1.1f%%') # Vẽ biểu đồ hình tròn
plt.title('Biểu đồ hình tròn')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [15, 16, 10, 18, 25]

plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')  # Vẽ biểu đồ phân tán
plt.title('Biểu đồ phân tán ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.xlabel('Trục x')  # Đặt nhãn cho trục x
plt.ylabel('Trục y')  # Đặt nhãn cho trục y
plt.legend()  # Hiển thị chú giải
plt.show()  # Hiển thị biểu đồ
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 10, 30, 35, 50, 50, 55, 65]
plt.boxplot(data)#vẽ biểu đồ
plt.title('Ví dụ biểu đồ hộp')#đặt tiêu đề cho biểu đồ
plt.ylabel('Giá trị')#đặt nhãn cho trục y
plt.show()
```

```python
import matplotlib.pyplot as plt
data = [15, 18, 22, 30, 35, 45, 50, 55, 65]
plt.violinplot(data)#vẽ biểu đồ hình violin
plt.title('Biểu đồ hình violin')#đặt tiêu đề
plt.ylabel('Giá trị')#đặt giá trị cho trục y
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [7, 2, 3, 5, 6]

N = len(categories)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Biểu đồ radar ví dụ')
plt.show()
```
